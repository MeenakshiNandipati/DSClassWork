### Making the train and test data ready for analysis

```python
# Making the train and test data ready for analysis
x_train = df_train[colnames[0:13]].values
y_train = df_train[colnames[13]].values

x_test = df_test[colnames[0:13]].values
y_test = df_test[colnames[13]].values
```

### Normalizing function

```python
##Normalmization
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

### Performing Multinomial Classification before Normalzing the data,
### It will be done after normalizing the data also.
### Looks like it performs better without normalizing the data.

```python
MultiNB = MultinomialNB ()
MultiNB. fit(x_train, y_train)
print(MultiNB)
y_pred = MultiNB.predict (x_test)
print (accuracy_score(y_test, y_pred))

print('Multinomial Classification type without Normalizing the data')
cm = confusion_matrix (y_test, y_pred)
cm
```

### Using the Normalize function on train and test data

```python
x_train = norm_func(x_train)
x_test =  norm_func(x_test)
```

## Applying all the 3 models on the data:
### Multinomial

```python
MultiNB = MultinomialNB ()
MultiNB. fit(x_train, y_train)
print(MultiNB)
y_pred = MultiNB.predict (x_test)
print (accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix (y_test, y_pred)
cm
```

```python
print('Multinomial Classification Report')
print(classification_report(y_test, y_pred))
```

```python
GausNB = GaussianNB ()
GausNB. fit (x_train, y_train)
print(GausNB)
y_pred = GausNB. predict (x_test)
print (accuracy_score (y_test, y_pred))
```

```python
cm = confusion_matrix (y_test, y_pred)
cm
```

```python
print('Gaussian Classification Report')
print(classification_report(y_test, y_pred))
```

```python
BernNB = BernoulliNB(binarize =0.1)
BernNB. fit (x_train, y_train)
print (BernNB)
y_pred = BernNB. predict (x_test)
print (accuracy_score (y_test, y_pred))
```
