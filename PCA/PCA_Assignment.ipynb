```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine1 = pd.read_csv("wine.csv")
wine = pd.read_csv("wine.csv")
wine.describe()
wine.head()
```

```python
# Ignoring the first column, which shows the type or cluster
wine.data = wine.iloc[:,1:]
wine.data.head()
# Converting into numpy array
wine = wine.data.values
wine
```

```python
# Normalizing the numerical data
wine_normal = scale(wine)
```

```python
wine_normal
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
#loadings or weights
pca.components_
```

```python
# The amount of variance that each PCA explains is
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained
plt.plot(var1,color="red")
```

```python
 pca_values[:,0:3]
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3']), wine1['Type']], axis = 1)
finalDf
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
import seaborn as sns
sns.scatterplot(data=finalDf,x='pc1',y='pc2',hue='Type',s = 100)
```

```python
p1 = sns.scatterplot(data=finalDf,x='pc1',y='pc2',s = 100)
for line in range(0,finalDf.shape[0]):
     p1.text(finalDf.pc1[line], finalDf.pc2[line], finalDf.Type[line], horizontalalignment='left', size='medium')
```

```python
# Visualization of PCAs
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=finalDf);
```

## For this data set, PCA did nto help much. Even after taking 3 components we got only 66% variance and 40% of the data is lost.

# Now trying other clusering models.
## Hierarchical clustering

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch # for creating dendrogram

z = linkage(wine_normal, method="ward",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=3, linkage='ward',affinity = "euclidean").fit(wine_normal)

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
wine1['Hclust']=cluster_labels # creating a  new column and assigning it to new column
wine1
```

```python
wine1['Hclust'].value_counts()
```

```python
wine1['Type'].value_counts()
```

# K mean Clustering on the same data

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(6, 4))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(wine_normal)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

Got the optimum value of k = 3 as per the elbow curve.

```python
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(wine_normal)
```

```python
cluster_labels=pd.Series(clf.labels_)
wine1['KClust']=cluster_labels # creating a  new column and assigning it to new column
wine1
```

```python
wine1['KClust'].value_counts()
```

```python
wine1.groupby(['Hclust','Type'])['Ash'].count()
```

As per the above group by function, we can see that Heirarchical clustering is more close to the original data. There are only a few mismatches.
