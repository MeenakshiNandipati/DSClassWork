# Apriori Algorithm

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
frequent_itemsets = apriori (book, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
print(len(frequent_itemsets))
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.4)
rules
rules.sort_values('lift',ascending = False).head(10)
```

#####  An leverage value of 0 indicates independence. Range will be [-1 1]
A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
movies = pd.read_csv("my_movies.csv")
movies.shape
```

```python
movies.head(20)
```

```python
#removing the first 5 columns, as the next columns have the respective dummy data.

movies= movies.iloc[:, 5:]
movies.head()
```

```python
frequent_itemsets = apriori (movies, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.4)
rules
rules.sort_values('lift',ascending = False).head(10)
```

```python
rules[rules.lift>1]
```
